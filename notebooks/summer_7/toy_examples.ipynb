## Correlated Planck reconstruct

```python
cov = np.array([
        [2.12e-08, -9.03e-08, 1.76e-08, 2.96e-07, 4.97e-07, 2.38e-07],
        [-9.03e-08, 1.39e-06, -1.26e-07, -3.41e-06, -4.15e-06, -3.28e-06],
        [1.76e-08, -1.26e-07, 9.71e-08, 4.30e-07, 7.41e-07, 4.13e-07],
        [2.96e-07, -3.41e-06, 4.30e-07, 5.33e-05, 9.53e-05, 1.05e-05],
        [4.97e-07, -4.15e-06, 7.41e-07, 9.53e-05, 2.00e-04, 1.35e-05],
        [2.38e-07, -3.28e-06, 4.13e-07, 1.05e-05, 1.35e-05, 1.73e-05]])
```

```python
planck_cov = np.linalg.eigvals(cov)
```

```python
samples = correlated_gaussian(nlive=500, mean=0.5*np.ones(6), cov=np.diag(planck_cov))
model = EndModel(samples)
true_endpoint = model.true_endpoint()
```

```python
iterations = np.linspace(0.05*true_endpoint, true_endpoint, 25, endpoint=False, dtype=int)
logXfs, logXfs_std = model.logXfs(theta_bandwidth_trunc, iterations, splits=1, Nset=25)
```

Now set the posterior for the yet unconstrained parameters to be constant

```python
trick_cov = np.linalg.eigvals(cov)
trick_cov[0] = 1e6
trick_cov[1] = 1e6
trick_cov[2] = 1e6
trick_cov[3] = 1e6
```

```python
samples_trick = correlated_gaussian(nlive=500, mean=0.5*np.ones(6), cov=np.diag(trick_cov))
model_trick = EndModel(samples_trick)
true_endpoint_trick = model_trick.true_endpoint()
```

```python
iterations_trick = np.linspace(0.05*true_endpoint_trick, true_endpoint_trick, 25, endpoint=False, dtype=int)
logXfs_trick, logXfs_std_trick = model_trick.logXfs(theta_bandwidth_trunc, iterations_trick, splits=1, Nset=25)
```

```python
logXs = samples.logX().iloc[iterations]
endpoints, endpoints_std = calc_endpoints(iterations, logXs, logXfs, logXfs_std, nlive=678)
logXs_trick = samples_trick.logX().iloc[iterations_trick]
endpoints_trick, endpoints_std_trick = calc_endpoints(iterations_trick, logXs_trick, logXfs_trick, logXfs_std_trick, nlive=678)
```

Plot predictions for the different samples

```python
fig, ax = plt.subplots(1, 1, figsize=(4,2))
plot_std(iterations, endpoints, endpoints_std, true_endpoint, ax=ax, ylim=(0, 1.5))
plot_std(iterations_trick, endpoints_trick, endpoints_std_trick, true_endpoint_trick, ax=ax)
```

```python
ndead = 2000
points = points_at_iteration(samples, ndead)[ndead:]
points_trick = points_at_iteration(samples_trick, ndead)[ndead:]
```

Both samples have the same likelihood profile at early iterations. 

Extrapolating from it leads to the right answer in one case, wrong answer in the other

```python
plt.plot(np.exp(samples.logX()), samples.logL)
plt.plot(np.exp(samples_trick.logX()), samples_trick.logL)
analytic_lm_params(points.logL, np.exp(points.logX()), d0=1)
```
