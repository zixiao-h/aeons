```python
# Generate plot of n_k
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({'font.size': 6})
```

```python
n = 500
n_dead = 1000
k = np.arange(0, n+n_dead)

n_k = np.zeros(n+n_dead)
n_k[:n_dead] = n
n_k[n_dead:] = n - (k[n_dead:] - n_dead)

plt.figure(figsize=(3,2), dpi=400)
plt.plot(k, n_k)
plt.axvline(x=n_dead, ymin=0, ymax = n/750, ls='--', lw=1)

plt.xlim(0,1600)
plt.ylim(0,750)
plt.xticks([0, 1000], [0, r'$n_{dead}$'])
plt.yticks([0, n], [None, 'n'])
plt.xlabel(r'$k$')
plt.ylabel(r'$n_k$')
plt.title(r'Distribution of live points $n_k$ through a run')
plt.savefig('nk_distribution.png')
```
