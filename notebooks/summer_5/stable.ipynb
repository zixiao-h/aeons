```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def characteristic_function(k):
    sigma = 1e-4
    t = sigma**2
    return np.exp(-k*t)
# Compute numerical Fourier transform
from scipy.fftpack import fft, fftfreq
# Number of sample points
N = 10000
# sample spacing
T = 10.0 / N
k = np.linspace(0.0, N*T, N, endpoint=False)
ck = characteristic_function(k)
px = fft(ck)
x = fftfreq(N, T)[:N//2]
plt.plot(x, np.log(2.0/N * np.abs(px[0:N//2])))
plt.xlim(0, 1)
```
