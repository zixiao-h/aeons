## Make results

```python
chain = 'gauss_32_2000'
name, samples = get_samples(chain)
model = EndModel(samples)
true_endpoint = model.true_endpoint()
true_logXf = model.true_logXf()
iterations = make_iterations(true_endpoint, 25, start=0.01)

d_Gs, d_Gs_std = model.d_Gs(get_d_G_prop, iterations)
logXfs, logXfs_std = model.logXfs(logXfs_prop, iterations, Nset=50)
write_to_txt(f'data/d_Gs/prop/{name}.txt', [iterations, d_Gs, d_Gs_std])
write_to_txt(f'data/logXfs/prop/{name}.txt', [iterations, logXfs, logXfs_std, true_endpoint])
```

```python
from joblib import Parallel, delayed

def process_chain(chain):
    name, samples = get_samples(chain)
    model = EndModel(samples)
    true_endpoint = model.true_endpoint()
    true_logXf = model.true_logXf()
    iterations = make_iterations(true_endpoint, 25, start=0.01)
    
    d_Gs, d_Gs_std = model.d_Gs(get_d_Gs_post, iterations)
    logXfs, logXfs_std = model.logXfs(logXfs_post, iterations, Nset=50)
    write_to_txt(f'data/d_Gs/post/{name}.txt', [iterations, d_Gs, d_Gs_std])
    write_to_txt(f'data/logXfs/post/{name}.txt', [iterations, logXfs, logXfs_std, true_endpoint])

chains = ['planck_gaussian', 'planck', 'gauss_16', 'gauss_32', 'gauss_64', 'cauchy_83', 'cauchy_85', 'lensing', 'BAO', 'correlated_6d']
n_jobs = -1  # Use all available CPU cores

# Parallelize the processing of chains
Parallel(n_jobs=n_jobs)(delayed(process_chain)(chain) for chain in chains)
```
